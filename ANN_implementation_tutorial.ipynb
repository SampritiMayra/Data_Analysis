```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Datasets/Churn_Modelling.csv')
df.head(10)
```

```python
# Divide the Dataset into independent and dependent features

X = df.iloc[:,3: 13]
Y= df.iloc[:, 13]
```

```python
X.head()
```

```python
Y
```

```python
# Feature Engineering
# Encoding Categorical Values

Geography = pd.get_dummies(X['Geography'])
Geography.head()
```

```python
Gender = pd.get_dummies(X['Gender'])
Gender.head()
```

```python
# concatenate these variables with the dataframe

X= X.drop(['Gender', 'Geography'], axis = 1)
```

```python
X = pd.concat([X,Geography,Gender], axis = 1)
X.head()
```

```python
from sklearn.model_selection import train_test_split #to split the dataset for training and testing
train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=0.2,random_state=0)
train_x.shape, train_y.shape
```

```python
# Feature scaling
# Usually in case of ANN, StandardScaler is used. Min-Max scaler is used in case of CNN
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(train_x)
X_test = sc.transform(test_x)
```

```python
# Now creating the ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, PReLU, ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
# Let's initialize the ANN 
model = Sequential()
```

```python
# Adding the input layer
model.add(Dense(units = 11, activation = 'relu'))
```

```python
# Adding the first hidden layer
model.add(Dense(units = 7, activation = 'relu'))
```

```python
# Adding the second hidden layer
model.add(Dense(units = 6, activation = 'relu'))
```

```python
# Adding the output layer
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# We're using Adam optimizer since it works the best 
# for loss function we're using binary_crossentropy because the problem we're solving is a binary classification.
model.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])
```

```python
# We can also set the learning rate
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.01)
#classifier.compile(optimizer = opt, loss ='binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
```

```python
# Implementation of early stopping and Saving best model
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)
```

```python
# Let's fit the model
model_history = model.fit(X_train, train_y, validation_split = 0.33, batch_size = 10,
                               epochs = 1000,callbacks = [es, mc])
```

```python
saved_model = load_model('best_model.h5')
```

```python
# Making the predictions and evaluating the model

y_pred = saved_model.predict(X_test)
y_pred = (y_pred >= 0.5)
```

```python
# Make the confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, y_pred)
cm
```

```python
# Calculate the accuracy 
from sklearn.metrics import accuracy_score
score = accuracy_score(test_y, y_pred)
score
```

```python
train_acc = saved_model.evaluate(train_x, train_y, verbose=1)
test_acc = saved_model.evaluate(test_x, test_y, verbose=1)
print('Train: %.3f, Test: %.3f', (train_acc, test_acc))
```
